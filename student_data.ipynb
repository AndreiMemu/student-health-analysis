## We first check the analysis with the estimate of 4hrs for the less than 5, and 9 for more than 8

```python
sleep_duration_approx = {
    'Less than 5 hours': 2,
    '5-6 hours': 5.5,
    '7-8 hours': 7.5,
    'More than 8 hours': 12,
    'Others': pd.NA
}

df['Sleep Duration'] = df['Sleep Duration'].map(sleep_duration_approx)

df
```

## Check what the population mean is and calculate the population variance and standard deviation

```python
mu = df['Sleep Duration'].sum() / len(df['Sleep Duration']) #  population mean

population_variance = ((df['Sleep Duration'] - mu)**2).sum() / len(df['Sleep Duration'])-1 # population variance

standard_deviation = population_variance ** 0.5 # standard deviation

print(f"The population mean of the sleep duration is: {round(mu, 2)} hours")

print(f"The standard deviation of the sleep duration is: {round(standard_deviation, 2)}")
```

Since we had all the data to refer to in this analysis, we could calculate the standard deviation. After all the calculations, we found that the standard deviation is in fact smaller than the population mean, which would result in the fact that most of our students do tend to sleep between 4.5 hours to 7-8 in this case of analysis.

We found that if we give the less than 5 hrs a value of 4 and to the more than 8 hrs a value of 9, the data shows the population mean to be 6.33hrs and the standard deviation to be 1.64
